# Using Pandas to explore College Major Data
This project will explore data about college majors and future earnings.  The data are taken from the 538 github repo.  They have cleaned the dataset, so I will spend very little time cleaning the data.  The github repo can be accessed here: https://github.com/fivethirtyeight/data/tree/master/college-majors

```python
import pandas as pd
import matplotlib as pt
%matplotlib inline

recent_grads= pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

# Plotting data with Pandas

I will explore relationships between different columns by plotting them using panda's plot method.

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Employed vs. Median', figsize=(5,10))
```

There doesn't seem to be a coorelation between students in more popular majors making more money.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full time vs. Median', figsize=(5,10))
```

There appears to be a slight negitive coorelation between full time and negitive salary.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs Unemployment Rate', figsize=(5,10))
```

There doesn't seem to be a coorelation between majoring in subjects that are majority female and making more money.  The plot looks to be random.

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(5,10))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs Median', figsize=(5,10))
```

# Exploring Distributions of different Columns

```python
recent_grads['Sample_size'].plot(kind='hist')
```

A large majority of majors have a sample size from 0-500.

```python
recent_grads['Median'].plot(kind='hist')
```

The median salaries for recent grads was most common between 20-40K.  There is a long tail towards the higher salary ranges.

```python
recent_grads['Employed'].plot(kind='hist')
```

The majority of colege grads were employed in smaller companies.

```python
recent_grads['Full_time'].plot(kind='hist')
```

This histogram shows a large tail for full time employment.  This could also mean that the sample size was small.

```python
recent_grads['ShareWomen'].plot(kind='hist')
```

This histogram shows a non-normal distribution.  There are a few majors with very low percentages of women.  But women seem to be fairly evenly distributed across majors.

```python
recent_grads['Unemployment_rate'].plot(kind='hist')
```

The unemployment rate shows a normal distribution centered around .5.  Meaning that most majors had about 50% of the former graduates unemployed.

```python
recent_grads['Men'].plot(kind='hist')
```

This shows that there are a lot of majors with a relitively low sample rate for men. 

```python
recent_grads['Women'].plot(kind='hist')
```

Like with the "Men" plot above, there tended to be a lot of majors with very few women answering.

# Creating Bar Plots in matplotlib

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

Women seem to be more represented in the last 10 rows than in the first 10.

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[-10:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

There doesn't seem to be a clear coorelation between major and unemployment rate.
